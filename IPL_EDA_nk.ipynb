# IPL HISTORY -----> ALL THE HISTORY AND STATISTICS OF THE IPL FROM 2008-2020

## Number of matches in each edition along with number of participating franchises and winners :

```python
franchise_teams=[]               # no_of_franchises [list]
season=[]                   # season [list]
matches_played = []         # no_of_matches_played  [list]
tournament_winner=[]               # winner  [list]
tournament_runnerup=[]               # runner_up  [list]

for y in df['Date'].dt.year.unique():
    season.append(y)
    
    teams_lis = df[df['Date'].dt.year==y].toss_winner.nunique()
    franchise_teams.append(teams_lis)
    
    match_played = df[df['Date'].dt.year==y].shape[0] 
    matches_played.append(match_played)
    
    winner_team = df[df['Date'].dt.year==y].tail(1).winner.unique()[0]
    tournament_winner.append(winner_team)
    
    if df[df['Date'].dt.year==y].tail(1).team1.unique()[0] == df[df['Date'].dt.year==y].tail(1).winner.unique()[0]:
        runner_up = df[df['Date'].dt.year==y].tail(1).team2.unique()[0]
        tournament_runnerup.append(runner_up)
    else:
        runner_up = df[df['Date'].dt.year==y].tail(1).team1.unique()[0]
        tournament_runnerup.append(runner_up)

teams_participated_dict = {'season':season,'no_of_matches_played':matches_played,'no_of_franchises':franchise_teams,'winner':tournament_winner,'runner_up':tournament_runnerup}
teams_participated = pd.DataFrame(teams_participated_dict)
teams_participated
```

## BOWLER STATS IN THE IPL 

```python
wickets=ipl_merged.query('(is_wicket!=0) and (dismissal_kind in ["caught","bowled","lbw","stumped","caught and bowled","hit wicket"])')
top_indiv_wickets = wickets.groupby(['bowler'])[['is_wicket']].count().sort_values(by='is_wicket',ascending=False).rename(columns={'is_wicket':'TOTAL_WICKETS'}).reset_index()



bowl_inn = []
overs_bowled = []
runs_conceded = []
for Bowler in top_indiv_wickets.bowler:
    inn_bowl = ipl_merged[ipl_merged['bowler']==Bowler].id.nunique()
    bowl_inn.append(inn_bowl)
    overs = round(ipl_merged.loc[(ipl_merged['bowler']==Bowler) & (ipl_merged['extras_type'] != 'wides') & (ipl_merged['extras_type'] != 'noballs') & (ipl_merged['extras_type'] != 'penalty')].shape[0]/6,2)
    overs_bowled.append(overs)
    runs = ipl_merged.loc[(ipl_merged['bowler']==Bowler) & (ipl_merged['extras_type'] != 'byes') & (ipl_merged['extras_type'] != 'legbyes') & (ipl_merged['extras_type'] != 'penalty') ].total_runs.sum()
    runs_conceded.append(runs)
top_indiv_wickets['no_of_innings'] = bowl_inn
top_indiv_wickets['overs_bowled']= overs_bowled
top_indiv_wickets['runs_conceded']= runs_conceded
top_indiv_wickets['AVERAGE'] = round(top_indiv_wickets['runs_conceded']/top_indiv_wickets['TOTAL_WICKETS'],2)
top_indiv_wickets['ECONOMY'] = round(top_indiv_wickets['runs_conceded']/top_indiv_wickets['overs_bowled'],2)
top_indiv_wickets['STRIKE_RATE'] = round((top_indiv_wickets['overs_bowled']*6)/top_indiv_wickets['TOTAL_WICKETS'],2)

top_35_wicket = top_indiv_wickets.head(35)

pd.set_option('display.max_rows',35)
top_indiv_wickets.head(35)

```

```python
# Visualizing Economy,Bowling Average and Strike rate of Top IPL bowlers in the history.
plt.figure (figsize= (16,12))
plt.rcParams['font.size'] = 16
sns.set_style('whitegrid')
plt.bar(top_35_wicket.bowler,top_35_wicket.AVERAGE,color='blue',width = 0.6,alpha=0.5)
plt.bar(top_35_wicket.bowler,top_35_wicket.STRIKE_RATE,color='yellow',width = 0.6,alpha=0.7)
plt.plot(top_35_wicket.bowler,top_35_wicket.ECONOMY,color='brown',marker='o',linewidth=0.2,markersize=6)
plt.xlabel('Bowler  --->')
plt.ylabel('Average / Economy / Strike Rate  --->')
plt.xticks(rotation=80)
plt.yticks(np.arange(4,38,1))
plt.legend(['Economy','Bowling Average','Bowling Strike Rate'],bbox_to_anchor=(1.25, 0.7))
plt.title("Top 35 IPL wicket takers' stats :")
plt.show()
plt.close();
```

## BATSMAN STATS IN THE IPL. 

```python
# top batsman with most runs in ipl
top_indiv_runs = ipl_merged.groupby('batsman')[['batsman_runs']].sum().sort_values(by='batsman_runs',ascending=False).rename(columns={'batsman_runs':'TOTAL_IPL_RUNS'}).reset_index()

bat_inn = []
outs = []
balls_faced= []
for Batsman in top_indiv_runs.batsman:
    inn_bat = ipl_merged[ipl_merged['batsman']==Batsman].id.nunique()
    bat_inn.append(inn_bat)
    out = ipl_merged[ipl_merged['player_dismissed']==Batsman].id.nunique()
    outs.append(out)
    ball_faced = ipl_merged.loc[(ipl_merged['batsman']==Batsman) & (ipl_merged['extras_type'] != 'wides')].ball.count() #no ball is counted in ball faced by batsman
    balls_faced.append(ball_faced)
not_outs = np.array(bat_inn) - np.array(outs)

top_indiv_runs['no_of_innings'] = bat_inn
top_indiv_runs['not_outs'] = not_outs
top_indiv_runs['AVERAGE'] = round(top_indiv_runs['TOTAL_IPL_RUNS']/outs,2)
top_indiv_runs['balls_faced'] = balls_faced
top_indiv_runs['SRTIKE_RATE'] = round((top_indiv_runs['TOTAL_IPL_RUNS']/balls_faced)*100,2)

top_35_indiv_runs = top_indiv_runs.head(35)

pd.set_option('display.max_rows',35)
top_indiv_runs.head(35)
```

```python
# Visualizing the Strike Rate and Average of Top Batsman in the IPL history
plt.figure (figsize= (18,11))
plt.rcParams['font.size'] = 16
sns.set_style('whitegrid')
plt.bar(top_35_indiv_runs.batsman,top_35_indiv_runs.SRTIKE_RATE,color='darkorange',width=0.6,alpha=0.77)
plt.plot(top_35_indiv_runs.batsman,top_35_indiv_runs.AVERAGE,color='maroon',marker='o',linewidth=0.2,markersize=7)
plt.xlabel('Batsman  --->')
plt.ylabel('Average / Strike Rate  --->')
plt.xticks(rotation=80)
plt.yticks(np.arange(15,165,5))
plt.legend(['Batting Average','Batting Strike Rate'],bbox_to_anchor=(1.2, 0.7))
plt.title("Top 35 IPL run scorers' stats :")
plt.show()
plt.close();
```

## FINALIST AND WINNER IN THE IPL HOSTORY

```python
year=[]            # season  [list]
finalists=[]       # finalist teams [list]
for y in ipl_matches.year.unique():
    for team in ipl_matches.team1.unique():
        x = ipl_matches[ipl_matches['year']==y].tail(1).team1.unique()[0]
        z = ipl_matches[ipl_matches['year']==y].tail(1).team2.unique()[0]
        
        if team in [x,z]:
            finalists.append(team)
            year.append(y)
        else:
            continue
finalist_dict = {'franchise':finalists,'Season':year}
finalist = pd.DataFrame(finalist_dict).groupby('franchise').count()[['Season']].rename(columns={'Season':'times_reached_finals'}).sort_values(by='times_reached_finals',ascending=False).reset_index()
finalist
```

# ORANGE CAPS AND PURPLE CAP STATS IN THE IPL HISTORY

* PURPLE CAP WINNERS

```python
wickets_new=ipl_merged.query('(is_wicket!=0) and (dismissal_kind in ["caught","bowled","lbw","stumped","caught and bowled","hit wicket"])')

name_of_bowler=[]
no_of_wickets=[]
season=[]
franchise1=[]
for y in wickets_new['Date'].dt.year.unique():
    season.append(y)
    bowler_name=wickets_new[wickets_new['Date'].dt.year==y].groupby(['bowler','bowling_team'])[['is_wicket']].count().sort_values(by='is_wicket',ascending=False).head(1).index[0][0]
    name_of_bowler.append(bowler_name)
    
    wickets_taken=wickets_new[wickets_new['Date'].dt.year==y].groupby(['bowler','bowling_team'])[['is_wicket']].count().sort_values(by='is_wicket',ascending=False).head(1).is_wicket[0]
    no_of_wickets.append(wickets_taken)
    
    team_bowling = wickets_new[wickets_new['Date'].dt.year==y].groupby(['bowler','bowling_team'])[['is_wicket']].count().sort_values(by='is_wicket',ascending=False).head(1).index[0][1]
    franchise1.append(team_bowling)
    
purple_cap_dict={'Season':season,'Purple_cap_winner':name_of_bowler,'No_of_wickets_taken':no_of_wickets,'Franchise':franchise1}

purple_cap = pd.DataFrame(purple_cap_dict)

purple_cap
```

* ORANGE CAP WINNER

```python
name_of_batsman=[]
no_of_runs=[]
season=[]
franchise1=[]
for y in ipl_merged['Date'].dt.year.unique():
    season.append(y)
    batsman_name=ipl_merged[ipl_merged['Date'].dt.year==y].groupby(['batsman','batting_team'])[['batsman_runs']].sum().sort_values(by='batsman_runs',ascending=False).head(1).index[0][0]
    name_of_batsman.append(batsman_name)
    
    runs_scored=ipl_merged[ipl_merged['Date'].dt.year==y].groupby(['batsman','batting_team'])[['batsman_runs']].sum().sort_values(by='batsman_runs',ascending=False).head(1).batsman_runs[0]
    no_of_runs.append(runs_scored)
    
    team_batting = ipl_merged[ipl_merged['Date'].dt.year==y].groupby(['batsman','batting_team'])[['batsman_runs']].sum().sort_values(by='batsman_runs',ascending=False).head(1).index[0][1]
    franchise1.append(team_batting)
    
orange_cap_dict={'Season':season,'Orange_cap_winner':name_of_batsman,'Total_runs':no_of_runs,'Franchise':franchise1}

orange_cap = pd.DataFrame(orange_cap_dict)

orange_cap
```
